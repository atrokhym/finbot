```python
import subprocess
import os

notebooks = ['Stock_NeurIPS2018_1_Data.ipynb',  'Stock_NeurIPS2018_2_Train.ipynb',  'Stock_NeurIPS2018_3_Backtest.ipynb'] #add other notebook names if there are more.

for notebook in notebooks:
    print(f"Executing {notebook}...")
    try:
        subprocess.run(['jupyter', 'nbconvert', '--to', 'notebook', '--execute', notebook, '--ExecutePreprocessor.timeout=None', '--output', f'{notebook.split(".")[0]}_executed.ipynb'], check=True)
        print(f"{notebook} executed successfully!")
    except subprocess.CalledProcessError as e:
        print(f"Error executing {notebook}: {e}")
```
